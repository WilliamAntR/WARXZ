```python
# Ambil semua maskapai unik
airlines = df['airline'].unique()

pyldavis_html = {}

for airline in airlines:
    print(f"\n=== Building LDA for {airline} ===")

    # Filter data hanya untuk maskapai ini
    airline_df = df[df['airline'] == airline]
    texts = airline_df['clean_text'].dropna().tolist()
    tokenized = [text.split() for text in texts]

    # Buat dictionary dan corpus
    id2word = Dictionary(tokenized)
    corpus = [id2word.doc2bow(text) for text in tokenized]

    # Buat LDA model
    lda_model = LdaModel(corpus=corpus,
                         id2word=id2word,
                         num_topics=4,
                         random_state=100,
                         passes=10,
                         alpha='auto')

    # Simpan visualisasi
    vis = gensimvis.prepare(lda_model, corpus, id2word)
    pyldavis_html[airline] = vis
```

```python
# Tampilkan semua visualisasi per airline
for airline in pyldavis_html:
    print(f"\n=== {airline} ===")
    display(pyldavis_html[airline])
```
